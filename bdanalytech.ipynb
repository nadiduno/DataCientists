# Cientistas

Gerando CPF aleatorios

```python
from faker import Faker
fake = Faker('pt_BR')
```

```python
def cpfcientista():
  cpf = fake.cpf()
  return cpf
```

Gerando Nomes aleatorios

```python
from gender_guesser.detector import Detector
detector = Detector()

def sexocientista(nome):
    genero = detector.get_gender(nome)
    if genero == 'female':
        return (1,'Feminino')
    elif genero == 'male':
        return (2,'Masculino')
    else:
        return (3,'Não definido')
```

```python
novo = {'CPF': [], 'nome': [], 'code_sexo': [], 'sexo': []}
for i in range(200):
  cpf=cpfcientista()
  nome=nome=fake.first_name()
  sexo=sexocientista(nome)
  nome2=nome + ' ' + fake.last_name()
  novo['CPF'].append(cpf)
  novo['nome'].append(nome2)
  novo['code_sexo'].append(sexo[0])
  novo['sexo'].append(sexo[1])

cientistanovo = pd.DataFrame(novo)
cientistanovo
```
